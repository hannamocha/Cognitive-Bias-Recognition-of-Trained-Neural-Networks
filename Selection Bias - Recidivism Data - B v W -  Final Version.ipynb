```python
from conx import Network
import random
```

```python
def getData(file):
    #open file
    tf = open(file, "r")
    content = tf.readlines()
    row_count = sum(1 for row in content)
    print(row_count)
    #Parse file
    input = []
    content = [x.strip() for x in content[1:]]
    for y in content:
        row = [col for col in y.split(",")]
        #print(row)
        input.append(row)
    tf.close()
    return input

cols = getData("compas-scores-two-years.csv")
```

```python
for c in cols:
    if len(c) == 54:
        c.remove(c[23])
        
print(len(cols[4062]))
```

```python
def parseInputs(inputs):
    new = []
    row = []
    countWhat = 0
    for r in inputs:
        if r[9] == "Caucasian" or r[9] == "African-American":
            #0
            if r[5] == "Male":
                row.append(1)
            elif r[5] == "Female":
                row.append(0)
            else:
                print("error parsing col 5, Error occurs on row " + str(countWhat))
            #Age: Greater than 45, 25 - 45, and Less than 25
            #1, 2, and 3
            if r[8] == "Greater than 45":
                row.append(1)
                row.append(0)
                row.append(0)
            elif r[8] == "25 - 45":
                row.append(0)
                row.append(1)
                row.append(0)
            elif r[8] == "Less than 25":
                row.append(0)
                row.append(0)
                row.append(1)
            else:
                print("error parsing col 8, Error occurs on row " + str(countWhat))
            #race = Caucasian, African-American
            if r[9] == "Caucasian":
                row.append(0)
            elif r[9] == "African-American":
                row.append(1)
            else:
                print("error parsing col 9, Error occurs on row " + str(countWhat))
            #Risk recidivism Score: Low, Medium, and High
            if r[40] == "Low":
                row.append(1)
                row.append(0)
                row.append(0)
            elif r[40] == "Medium":
                row.append(0)
                row.append(1)
                row.append(0)
            elif r[40] == "High":
                row.append(0)
                row.append(0)
                row.append(1)
            else:
                print("error parsing col 40, Error occurs on row " + str(countWhat))
            #Risk of violence Score: Low, Medium, and High
            if r[44] == "Low":
                row.append(1)
                row.append(0)
                row.append(0)
            elif r[44] == "Medium":
                row.append(0)
                row.append(1)
                row.append(0)
            elif r[44] == "High":
                row.append(0)
                row.append(0)
                row.append(1)
            else:
                print("error parsing col 44, Error occurs on row " + str(countWhat))
            #If recidivism
            if r[52] == '0':
                row.append(0)
            elif r[52] == '1':
                row.append(1)
            else:
                print("error parsing col 52, Error occurs on row " + str(countWhat))
            
            #parse error in row 4754, disposed of here
            if r[40] == "":
                r = []

            if row != [] and len(row) == 12:
                new.append(row)
            #reset
            row = []
        countWhat += 1
    return new

tempinputs = parseInputs(cols)  
tempinputs[0]
```

```python
#Ensure inputs are coorect length
#Input how many rows to be processed
inputs = []
#lengthInput = int(.8 * len(tempinputs))
lengthInput = 4000
print("Total length of inputs: " + str(lengthInput))

for r in range(lengthInput):
    inputs.append(tempinputs[r])
    
inputs = [[x[0:11], x[11:]] for x in inputs]
```

```python
inputs[0]
```

```python
net = Network(11, 60, 1)

#desired output/target
def des(inputs):
    re = inputs[12]
    return [int(re)]
```

```python
net.set_inputs(inputs)
```

```python
net.reset()
#net.train(max_training_epochs=1000, momentum=1, epsilon=1)
net.train(max_training_epochs=1500, report_rate=250, tolerance = .4)
```

```python
net.test()
```

```python
wresults = []
bresults = []
wtest = []
btest = []

for p in range(len(tempinputs) - 4000):
    val = tempinputs[p+4000]
    #White
    if val[4] == 0:
        wtest.append(val[0:11])
    #black
    if val[4] == 1:
        btest.append(val[0:11])
    
for l in wtest:
    wresults.append(net.propagate(l))
    
for m in btest:
    bresults.append(net.propagate(m))

wFalseCount = 0
bFalseCount = 0
falseCountTrainData = 0

for wr in wresults:
    if wr < .5:
        wFalseCount += 1

for br in bresults:
    if br < .5:
        bFalseCount += 1
        
for rowz in tempinputs:
    if rowz[11:] == [0]:
        falseCountTrainData += 1

print("\n")  
print("wFalseCount")   
print(wFalseCount)

print("\n")  
print("bFalseCount")   
print(bFalseCount)
        
print("falseCountTrainData")   
print(falseCountTrainData)
```

```python
def getBias(falseCount, testSet):
    numer = falseCount/len(testSet)
    demo = falseCountTrainData/lengthInput
    print("Bias ratio given test dataset: ")
    return numer/demo
```

```python
print("White cases: ")
print(getBias(wFalseCount, wtest))
```

```python
print("Black Cases: ")
print(getBias(bFalseCount, btest))
```
